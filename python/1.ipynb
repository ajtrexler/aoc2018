```python
with open("../data/1.txt") as f:
    data = f.read()
```

```python
# part one
num = 0
for d in data.split("\n"):
    if "-" in d:
        step = int(d.split("-")[1])
        num = num - step
    elif "+" in d:
        step = int(d.split("+")[1])
        num = num + step
print(num)
```

```python
# part two
data = "-6, +3, +8, +5, -6"
data = data.replace(", ","\n")
data
```

```python
num = 0
seen = set()
keep_going = True
while keep_going:
    for d in data.split("\n"):
        if "-" in d:
            step = int(d.split("-")[1])
            num = num - step
        elif "+" in d:
            step = int(d.split("+")[1])
            num = num + step
        if num not in seen:
            seen.add(num)
        else:
            print(num)
            keep_going = False
            break
```
